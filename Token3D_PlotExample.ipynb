```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
words = ['Man', 'Woman', 'Girl', 'Baby', 'Sir', 'Bro']
coordinates = [
    [-1, 1, 1, 0, -1, -1],
    [0, 0, 0.5, -1, 0.5, -0.5],
    [0, 0, 0, 0, 1, -1]
]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points
for i, word in enumerate(words):
    x, y, z = coordinates[0][i], coordinates[1][i], coordinates[2][i]
    ax.scatter(x, y, z, color='b')
    ax.text(x, y, z, word, fontsize=12)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show plot
plt.show()
```
